```python
# ecoregions_gdf and occurrences_df would not join because there are no common column names
# change OBJECTID in ecoregion_gdf to 'ecoregion'
ecoregions_gdf = ecoregions_gdf.rename(columns={
    'OBJECTID':'ecoregion_id'
})
ecoregions_gdf.head()
```

```python
# set ecoregion_gdf index to ecoregion
ecoregions_gdf.set_index('ecoregion_id', inplace=True)
```

```python
occurrence_df.head()
```

```python
# Join the occurrences with the plotting GeoDataFrame
occurrence_gdf = ecoregions_gdf.join(occurrence_df)
occurrence_gdf.head()
```

```python


# Get the plot bounds so they don't change with the slider
xmin, ymin, xmax, ymax = occurrence_gdf.total_bounds

# Month slider widget
slider = pn.widgets.DiscreteSlider(
        options={
            calendar.month_name[month_num]: month_num
            for month_num in range(1,13)
        }
)

# Plot occurrence by ecoregion and month
migration_plot = (
    occurrence_gdf
    .hvplot(
        c='norm_occurrences',
        groupby='month',
        # Use background tiles
        geo=True, crs=ccrs.Mercator(), tiles='CartoLight',
        title="Veery Migration Patterns",
        xlim=(xmin, xmax), ylim=(ymin, ymax),
        frame_height=600,
        widgets={'month': slider},
        widget_location='bottom'
    )
)



# Show the plot
migration_plot
```

```python
# Save the plot
migration_plot.save('migration.html', embed=True)
```
